```python
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = load_diabetes()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['target'] = dataset.target

# Pairplot
sns.pairplot(df[['bmi','bp','s1','s5','target']])
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="viridis")
plt.title("Healthcare Data Correlation Heatmap")
plt.show()

# Scatter plot
sns.scatterplot(x='bmi', y='target', data=df)
plt.title("BMI vs Disease Progression")
plt.show()
```

```python
# !pip install folium pandas

import pandas as pd
import folium

data = {
    'City': ['New York', 'London', 'Tokyo', 'Sydney', 'Cairo'],
    'Latitude': [40.7128, 51.5074, 35.6895, -33.8688, 30.0444],
    'Longitude': [-74.0060, -0.1278, 139.6917, 151.2093, 31.2357],
    'Population': [8.4, 9.0, 14.0, 5.3, 9.8] # Population in millions
}
df = pd.DataFrame(data)

print("Sample DataFrame:")
print(df)

world_map = folium.Map(location=[20, 0], zoom_start=2, tiles='CartoDB positron')

for index, row in df.iterrows():
    folium.Marker(
    location=[row['Latitude'], row['Longitude']],
    popup=f"<b>City:</b> {row['City']}<br><b>Population:</b> {row['Population']} million", tooltip=row['City']
    ).add_to(world_map)

world_map
```
