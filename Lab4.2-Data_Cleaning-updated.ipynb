```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "Age": [25, None, 42, 29, 42, 30], # None = missing value
    "Birthday": ["1998-05-14", "1980-07-09", "2010-03-07", "1994-11-21", "1981-12-30", "1993-08-19"],
    "Salary": [50000, 60000, -10, 45000, 70000, 50000], # -10 = noisy/invalid
    "Rating": ["A", "B", "3", "2", "B", "A"], # Mixed rating systems (inconsistent)
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# 1. Detect missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# 2. Detect noisy data (negative salary)
noisy_salary = df[df["Salary"] < 0]
print("\nNoisy salary rows:")
print(noisy_salary)

# 3. Detect inconsistent ratings
print("\nUnique ratings found:")
print(df["Rating"].unique())

# 4. Example cleaning steps
df["Salary"] = df["Salary"].apply(lambda x: np.nan if x < 0 else x) # Replace negative with NaN
df["Age"] = df["Age"].fillna(df["Age"].mean()) # Replace missing age with mean
df["Rating"] = df["Rating"].replace({"1": "A", "2": "B", "3": "C"}) # Standardize ratings

print("\nCleaned Data:")
print(df)
```

```python
data = {
    "ID": [1, 2, 3, 4, 5],
    "Income": [56000, np.nan, 45000, np.nan, 67000],
    "CreditRisk": ["High", "Low", "High", "Low", "High"]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# 1. Ignore tuple (drop rows with missing values)
ignore_tuple = df.dropna()
print("\n1. Ignore Tuple:\n", ignore_tuple)

# 2. Fill manually (example: setting income of row 2 as 50000)
manual_fill = df.copy()
manual_fill.loc[1, "Income"] = 50000
print("\n2. Fill Manually:\n", manual_fill)

# 3. Use a global constant (e.g., -1)
global_const = df.copy()
global_const["Income"] = global_const["Income"].fillna(-1)
print("\n3. Global Constant:\n", global_const)

# 4a. Use mean of the attribute
mean_fill = df.copy()
mean_fill["Income"] = mean_fill["Income"].fillna(df["Income"].mean())
print("\n4. Mean Fill:\n", mean_fill)

# 4b. Use median of the attribute
median_fill = df.copy()
median_fill["Income"] = median_fill["Income"].fillna(df["Income"].median())
print("\n4. Median Fill:\n", median_fill)

# 5. Use most probable value (mode)
mode_fill = df.copy()
mode_val = df["Income"].mode()[0]
mode_fill["Income"] = mode_fill["Income"].fillna(mode_val)
print("\n5. Mode (Most Probable) Fill:\n", mode_fill)
```

```python
from datetime import datetime

df = pd.read_csv("content/netflix_titles.csv")

# 1.
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')

# 2.
df['country'] = df['country'].replace(
    {"U.S.": "United States", "USA": "United States"}
)

# 3.
df = df.drop_duplicates()

# 4.
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# 5.
df = df.drop(columns=['show_id'])

# 6.
df = df.drop_duplicates(subset=['title', 'release_year'])

# 7.
current_year = datetime.now().year
df['content_age'] = current_year - df['release_year']

print(df.head())
```
