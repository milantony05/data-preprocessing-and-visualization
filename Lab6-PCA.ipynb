```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
breast_data = breast.data
print(breast_data)
print(breast_data.shape)
```

```python
breast_lables=breast.target
print(breast_lables)
print(breast_lables.shape)
```

```python
labels = np.reshape(breast_lables,(569,1))
final_breast_data = np.concatenate([breast_data,labels],axis=1)
print(final_breast_data.shape)
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
print(breast_dataset.head())
```

```python
features = breast.feature_names
print(features)
```

```python
# get feature names
columns = list(breast.feature_names) + ["label"]
# create dataframe with column names
breast_dataset = pd.DataFrame(final_breast_data, columns=columns)
print(breast_dataset.head())
```

```python
features_labels = np.append(features,'label')
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)
breast_dataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
print(x.shape)
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
print(normalised_breast)
```

```python
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_Df = pd.DataFrame(data = principalComponents_breast, columns = ['principal component 1', 'principal component 2'])
principal_breast_Df.tail()
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = [0,1]
colors = ['r', 'g']
#labels_map = {0: 'Malignant', 1: 'Benign'}
for target, color in zip(targets,colors):
    indicesToKeep = breast_lables == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1'], principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)
plt.legend(targets,prop={'size': 15})
```

```python
# Import required libraries
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Load the Diabetes dataset
# This dataset has 10 features and a continuous target (disease progression)
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = diabetes.target # Continuous target variable

# Scale features for better performance of regression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y,
test_size=0.3, random_state=42)

# Initialize Linear Regression model
model = LinearRegression()

# ---------------- Forward Feature Selection ----------------
# Select the top 5 features using forward selection
# Forward selection starts with zero features and adds the best feature iteratively
sfs_forward = SequentialFeatureSelector(model, n_features_to_select=5,
direction='forward')
sfs_forward.fit(X_train, y_train)

# Features selected by forward selection
forward_features = X.columns[sfs_forward.get_support()]
print("Forward Selection - Selected Features:",
list(forward_features))

# Train model using selected features
model.fit(X_train[:, sfs_forward.get_support()], y_train)
y_pred_forward = model.predict(X_test[:, sfs_forward.get_support()])
mse_forward = mean_squared_error(y_test, y_pred_forward)
print("Forward Selection - Mean Squared Error:", mse_forward)

# ---------------- Backward Elimination ----------------
# Select top 5 features using backward elimination
# Backward elimination starts with all features and removes the least useful iteratively
sfs_backward = SequentialFeatureSelector(model,
n_features_to_select=5, direction='backward')
sfs_backward.fit(X_train, y_train)

# Features selected by backward elimination
backward_features = X.columns[sfs_backward.get_support()]
print("Backward Elimination - Selected Features:",
list(backward_features))

# Train model using selected features
model.fit(X_train[:, sfs_backward.get_support()], y_train)
y_pred_backward = model.predict(X_test[:, sfs_backward.get_support()])
mse_backward = mean_squared_error(y_test, y_pred_backward)
print("Backward Elimination - Mean Squared Error:", mse_backward)
```
